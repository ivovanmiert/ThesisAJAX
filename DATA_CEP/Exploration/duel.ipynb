```python
import os
import ffmpeg
from datetime import timedelta
import yaml
import pandas as pd
import numpy as np
from itertools import combinations
from itertools import product
```

```python
df = pd.read_csv('D:\\GITHUB\\Data_CEP\\Exploration\\dataframe_expanded.csv')
```

```python
df_duel = df[df['primary_type'] == 'duel']
df_duel
```

```python
pd.set_option('display.max_rows', 500)
df_counts = df_duel['secondary_types'].value_counts()
df_counts
```

```python
condition1 = (df_duel['aerial_duel'] == 1)
condition2 = (df_duel['defensive_duel'] == 1)
condition3 = (df_duel['loose_ball_duel'] == 1)

# Willekeurig 2000 rijen selecteren voor elke voorwaarde
sample1 = df_duel[condition1].sample(n=2000, random_state=1)
sample2 = df_duel[condition2].sample(n=2000, random_state=1)
sample3 = df_duel[condition3].sample(n=2000, random_state=1)

# Combineer de gesamplede DataFrames
df_combined = pd.concat([sample1, sample2, sample3], ignore_index=True)

# Shuffle de gecombineerde DataFrame
df_selected = df_combined.sample(frac=1, random_state=1)

# Schrijf de geshufflede DataFrame naar een CSV-bestand
df_selected.to_csv('D:\\GITHUB\\Data_CEP\\Exploration\\selected_clips_duel.csv', index=False)
```
