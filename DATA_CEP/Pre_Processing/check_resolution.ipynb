```python
import os
import ffmpeg
```

```python
def get_video_resolution(video_path):
    try:
        probe = ffmpeg.probe(video_path)
        video_stream = next((stream for stream in probe['streams'] if stream['codec_type'] == 'video'), None)
        if video_stream:
            width = video_stream['width']
            height = video_stream['height']
            return width, height
        return None
    except ffmpeg.Error as e:
        print(f"Error occurred while probing video {video_path}: {e}")
        return None
```

```python
def list_videos_with_resolution(folder_path):
    video_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.mp4', '.mkv', '.avi', '.mov', '.wmv'))]
    video_info = []
    for video_file in video_files:
        video_path = os.path.join(folder_path, video_file)
        resolution = get_video_resolution(video_path)
        video_info.append((video_file, resolution))
    return video_info
```

```python
folder_path = 'D:\\Footage\\All_matches'
video_info_list = list_videos_with_resolution(folder_path)

for video_file, resolution in video_info_list:
    if resolution:
        print(f"{video_file}: {resolution[0]}x{resolution[1]}")
    else:
        print(f"{video_file}: Resolution not found")
```
