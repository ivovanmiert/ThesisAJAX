```python
import cv2
import os
```

```python
def extract_frames(video_path, output_folder, start_frame, end_frame):
    # Create output directory if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Open the video file
    video_capture = cv2.VideoCapture(video_path)

    # Set the video to start at the start_frame
    video_capture.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    current_frame = start_frame
    print(current_frame)

    while current_frame <= end_frame:
        # Read the frame at the current position
        ret, frame = video_capture.read()

        # Break the loop if no frame is returned (end of video)
        if not ret:
            print(f"End of video reached at frame {current_frame}")
            break

        # Save the frame as an image file
        output_filename = os.path.join(output_folder, f"frame_{current_frame}.jpg")
        cv2.imwrite(output_filename, frame)
        print(f"Saved {output_filename}")

        current_frame += 1

    # Release the video capture object
    video_capture.release()
    print("Frame extraction complete.")
```

```python
# Example usage
video_path = "D:\\Footage\\All_matches\\g5476244-hd.mp4"
output_folder = "D:\\GITHUB\\Data_CEP\\Pre_Processing\\Slicing\\output_test"
start_frame = 25000  # Replace with your starting frame number
end_frame = 25250   # Replace with your ending frame number
```

```python
extract_frames(video_path, output_folder, start_frame, end_frame)
```
