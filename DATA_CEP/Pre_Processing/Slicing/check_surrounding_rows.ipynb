```python
import pandas as pd
```

```python
df_full = pd.read_csv('D:\\GITHUB\\Data_CEP\\Exploration\\dataframe_expanded.csv')
```

```python
def get_surrounding_rows(df, event_id, n=4):
    # Find the index of the row with the specific event_id
    idx = df.index[df['event_id'] == event_id].tolist()
    
    if not idx:
        raise ValueError(f"event_id {event_id} not found in the DataFrame")
    
    idx = idx[0]
    
    # Calculate the start and end indices
    start_idx = max(idx - n, 0)
    end_idx = min(idx + n + 1, len(df))
    
    # Return the slice of the DataFrame
    return df.iloc[start_idx:end_idx]
```

```python
event_id_to_find = 1818891402
surrounding_rows = get_surrounding_rows(df_full, event_id_to_find)
print(surrounding_rows) 
```
